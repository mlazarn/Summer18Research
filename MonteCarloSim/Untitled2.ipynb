```python
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import argparse as ap
from scipy.fftpack import fft
from scipy.fftpack import fftfreq
from scipy.stats import linregress as linregress
import matplotlib.animation as manimation

plt.rc('text', usetex=True)
```

```python
data1 = np.genfromtxt("data/autoCorrTest11/size_0/autoCorr_10000.csv", delimiter=',')
data2 = np.genfromtxt("data/autoCorrTest11/size_1/autoCorr_10000.csv", delimiter=',')
data3 = np.genfromtxt("data/autoCorrTest11/size_2/autoCorr_10000.csv", delimiter=',')
data4 = np.genfromtxt("data/autoCorrTest11/size_3/autoCorr_10000.csv", delimiter=',')
for i in range(10100, 30000, 100):
    data1 = data1 + np.genfromtxt("data/autoCorrTest11/size_0/autoCorr_{}.csv".format(i), delimiter=',')
    data2 = data2 + np.genfromtxt("data/autoCorrTest11/size_1/autoCorr_{}.csv".format(i), delimiter=',')
    data3 = data3 + np.genfromtxt("data/autoCorrTest11/size_2/autoCorr_{}.csv".format(i), delimiter=',')
    data4 = data4 + np.genfromtxt("data/autoCorrTest11/size_3/autoCorr_{}.csv".format(i), delimiter=',')
data1 = data1/300
data2 = data2/300
data3 = data3/300
data4 = data4/300
```

```python
r = np.arange(data2.shape[1])
```

```python
fig1, ax1 = plt.subplots(facecolor="white")
ax1.plot(r[1:128], data1[80,1:128], label="Height = 256")
ax1.plot(r[1:128], data2[80,1:128], label="Height = 512")
ax1.plot(r[1:128], data3[80,1:128], label="Height = 640")
ax1.plot(r[1:128], data4[80,1:128], label="Height = 796")
ax1.axhline(0, linestyle='--', color='k')
ax1.set_xlabel(r'$r$')
ax1.set_ylabel(r'$\langle C(r)\rangle$')
#ax1.set_ylim(bottom=-0.05, top=0.125)
ax1.legend()
```

```python
fig2, ax2 = plt.subplots(facecolor="white")
ax2.plot(r[1:128], np.log(data1[80,1:128]), label="Height = 256")
ax2.plot(r[1:128], np.log(data2[80,1:128]), label="Height = 512")
ax2.plot(r[1:128], np.log(data3[80,1:128]), label="Height = 640")
ax2.plot(r[1:128], np.log(data4[80,1:128]), label="Height = 796")
ax2.axhline(0, linestyle='--', color='k')
ax2.set_xlabel(r'$r$')
ax2.set_ylabel(r'$\langle C(r)\rangle$')
#ax2.set_ylim(bottom=-0.05, top=0.125)
ax2.legend()
```

```python
x = np.arange(1, 21)
line1 = linregress(x, np.log(data1[80,1:21]))
line2 = linregress(x, np.log(data2[80,1:21]))
line3 = linregress(x, np.log(data3[80,1:21]))
line4 = linregress(x, np.log(data4[80,1:21]))
```

```python
syst_size = np.array([256, 512, 640, 796])
corr_len = 1/np.abs(np.array([line1.slope, line2.slope, line3.slope, line4.slope]))
```

```python
fig3, ax3 = plt.subplots(facecolor="white")
ax3.plot(syst_size, corr_len, label="Height = 256")
```
