```python
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
import argparse as ap
from scipy.fftpack import fft
from scipy.fftpack import fftfreq
import matplotlib.animation as manimation

plt.rc('text', usetex=True)
```

```python
data2 = np.genfromtxt("data/autoCorrTest5/autoCorr_1000.csv", delimiter=',')
for i in range(1010, 5000, 10):
    data2 = data2 + np.genfromtxt("data/autoCorrTest5/autoCorr_{}.csv".format(i), delimiter=',')

data2 = data2 / 500
```

### Vertical Auto-correlation function in ML bulk (interface + 244)

```python
#data2 = data2/3
r2 = np.arange(data2.shape[1]) + 1
fig1, ax1 = plt.subplots(facecolor="white")
ax1.plot(r2[0:128], data2[288,0:128])
ax1.set_xlabel(r'$r$')
ax1.set_ylabel(r'$\langle C(r)\rangle$')
ax1.set_ylim(bottom=-0.05, top=0.1)
```

```python
fig10, ax10 = plt.subplots(facecolor="white")
ax10.plot(r2[1:128], np.log(data2[288,1:128]))
#ax10.set_xlabel(r'$r$')
#ax10.set_ylabel(r'$\langle C(r)\rangle$')
#ax10.set_ylim(bottom=-0.05, top=0.1)
```

```python
specdat1 = fft(data2[288,1:128])
freq = fftfreq(127)
fig8, ax8 = plt.subplots(facecolor="white")
ax8.plot(freq[0:64], specdat1[0:64], 'k.')
#ax1.set_xlabel(r'$r$')
#ax1.set_ylabel(r'$\langle C(r)\rangle$')
#ax1.set_ylim(bottom=-0.05, top=0.1)
```

```python
maximum = np.max(specdat1[0:64])
argmax = np.argmax(specdat1[0:64])
1/freq[argmax]
```
